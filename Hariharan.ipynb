```python
import pandas as pd
# load the dataset
dataset = pd.read_csv('french_dictionary.csv',header=None)
dataset
```

```python
en_fr={}
for idx,row in dataset.iterrows() :
    en_fr.setdefault(row[0], row[1])

en_fr        
```

```python
ObjRead = open("t8.shakespeare.txt", "r")
 
 
 
Content = ObjRead.read();

 
ObjRead.close()
```

```python
Single_Lined= Content.split()
Single_Lined
```

```python
changed = []
for word in Single_Lined:
    try:
        k = word.lower()
        if en_fr[k] or en_fr[k][:-2]:
            if k[-2:] == "'s":
                k = word.lower()[:-2]
                if word.isupper():
                    changed.append(en_fr[k].upper() + "'s")
                    print("Upper ",en_fr[k].upper() + "'s")
                elif word.islower():
                    changed.append(en_fr[k].lower() + "'s")
                    print("Lower ",en_fr[k].lower() + "'s")
                elif word.istitle():
                    changed.append(en_fr[k].title() + "'s")
                    print("Title ",en_fr[k].title() + "'s")
            else:
                if word.isupper():
                    changed.append(en_fr[k].upper())
                elif word.islower():
                    changed.append(en_fr[k].lower())
                elif word.istitle():
                    changed.append(en_fr[k].title())

    except Exception as e:
        changed.append(word)
        pass
```
